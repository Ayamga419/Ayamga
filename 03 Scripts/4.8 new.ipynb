```python
ords_prod_merge['max_order'] = ords_prod_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prod_merge.head(15)
```

```python
ords_prod_merge.head(100)
```

```python
pd.options.display.max_rows = None
```

```python
ords_prod_merge.head(100)
```

```python
ords_prod_merge.loc[ords_prod_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prod_merge.loc[(ords_prod_merge['max_order'] <= 40) & (ords_prod_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prod_merge.loc[ords_prod_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prod_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prod_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
ords_prod_merge['average_spending'] = ords_prod_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prod_merge.head(15)
```

```python
ords_prod_merge.loc[ords_prod_merge['average_spending'] < 10, 'spending_flag'] = "low spender"
```

```python
ords_prod_merge.loc[ords_prod_merge['average_spending'] >= 10, 'spending_flag'] = "high spender"
```

```python
ords_prod_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Creating new column 'average_spending'

ords_prod_merge['purchase_frequency'] = ords_prod_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prod_merge.head(15)
```

```python
# Creating frequency flag - 'Non-frequent customer'

ords_prod_merge.loc[ords_prod_merge['purchase_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating frequency flag - 'Regular customer'

ords_prod_merge.loc[(ords_prod_merge['purchase_frequency'] > 10) & (ords_prod_merge['purchase_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Creating frequency flag - 'Frequent customer'

ords_prod_merge.loc[(ords_prod_merge['purchase_frequency'] <= 10), 'frequency_flag'] = 'Frequent Customer'
```

```python
ords_prod_merge['frequency_flag'].value_counts(dropna = False)
```
